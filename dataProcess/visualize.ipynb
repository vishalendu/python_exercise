```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

sns.set_style('darkgrid')
df = pd.read_csv('./data/output/pivot1.csv',index_col='ID')
fig,ax = plt.subplots(figsize=(10,10))
plt.title('Heatmap-Weather')
ax = sns.heatmap(df,mask=df.isnull())
plt.yticks(rotation=90)
fig.savefig('./data/output/pivot.png',dpi=500)
```
