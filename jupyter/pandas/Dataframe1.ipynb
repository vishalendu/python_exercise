## Using Tips dataset from here

```python
tip = sns.load_dataset('tips')
tip
```

```python
# order by smoker and sex and generate mean of total_bill and tip then delete one column

tot_bill = tip.groupby(['smoker','sex'])[['total_bill','tip']].sum()
tip1 = tip.groupby(['smoker','sex'])[['total_bill','tip']].sum()

del tot_bill['tip']
del tip1['total_bill']

tot_bill,tip1
```

```python
pd.merge(tot_bill.reset_index(),tip1.reset_index())
```

```python
tot_bill.reset_index().head(2)
```

```python
tip1.reset_index().tail(2)
```

```python
pd.merge(tot_bill.reset_index(),tip1.reset_index())
```

```python
pd.merge(tot_bill.reset_index(),tip1.reset_index().head(2),how='inner',indicator=True)
```

```python
pd.merge(tot_bill.reset_index().tail(3),tip1.reset_index().head(3),how='left',indicator=True)
```

```python
pd.merge(tot_bill.reset_index().tail(3),tip1.reset_index().head(3),how='right',indicator=True)
```

```python
pd.merge(tot_bill.reset_index().head(2),tip1.reset_index().tail(2),how='outer')
```

```python
pd.merge(tot_bill.reset_index().head(2),tip1.reset_index(),how='outer',indicator=True)
```

```python
pd.merge(tot_bill,tip1,left_index=True,right_index=True)
```

## Cleaning up data 

```python
nba.dropna() # removes rows with any column valus NaN or null values
```

```python
nba.dropna(how='all') ## remove rows with all columns values NaN
```

```python
nba.dropna(how='all',inplace=True)
```

```python
# Removes any column with Null Values
nba.dropna(axis=1) # default how = any, so if any value in the column is null it will be dropped.
```

```python
# if you only need to check for Nulls in one column
# this will remove rows where this column is null
nba_sal_nonull = nba.dropna(subset=['Salary'])
```

```python
# if you want to get a diff between multiple dataframes
nba_nulls = pd.concat([nba,nba_sal_nonull]).drop_duplicates(keep=False)
```

```python
nba_nulls
```

```python
# changig column level NaN to specific default values
nba['Salary'].fillna(0,inplace=True)
nba['College'].fillna('No College',inplace=True)
```

## Astype example

```python
# load and clean dataframe
nba = pd.read_csv('data/nba.csv').dropna(how='all')
nba['Salary'].fillna(0,inplace=True)
nba['College'].fillna('No College',inplace=True)
```

```python
nba
```

```python
nba.info()
```

```python
# change data types for some columns
for col in ['Salary','Age','Number']:
    nba[col] = nba[col].astype('int')
```

```python
nba.info()
```

```python
nba['Position'].nunique()
nba['College'].nunique()
nba['Team'].nunique()
```

```python
for col in ['Team','Position','College']:
    nba[col] = nba[col].astype('category')
```

```python
nba.info()
```

```python
# sorting with multiple columns with mixed sorting order
nba.sort_values(['Team','Name','Salary'],ascending=[True,True,False])
```

```python
nba.sort_index(axis=1,ascending=True)
```

```python
nba['SalaryRank'] = nba['Salary'].rank(ascending=False).astype('int')
```

```python
nba.sort_values('SalaryRank').head()
```
