```python
chicago = pd.read_csv('data/chicago.csv')
chicago.dropna(how='all',inplace=True)
chicago['LName'] = chicago['Name'].str.split(',').str.get(0)
chicago['FName'] = chicago['Name'].str.split(',').str.get(1).str.strip().str.split(' ').str.get(0)
chicago['MName'] = chicago['Name'].str.split(',').str.get(1).str.strip().str.split(' ').str.get(1)
for col in ['Name','FName','LName','Position Title','Department']:
    chicago[col]=chicago[col].str.title()
chicago['Department']  = chicago['Department'].astype('category')
chicago['FName']  = chicago['FName'].astype('category')
chicago['MName']  = chicago['MName'].astype('category')
chicago['Employee Annual Salary'] = chicago['Employee Annual Salary'].str.replace('$','')
chicago['Employee Annual Salary'] = pd.to_numeric(chicago['Employee Annual Salary'],\
                                                  errors='coerce')
chicago.columns = chicago.columns.str.replace(' ','_')
```
