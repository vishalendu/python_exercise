```python
fortune = pd.read_csv("data/fortune1000.csv")
```

```python
fortune.groupby(["Sector"]).agg({"Revenue":"sum","Profits":"sum","Employees":"sum"}).\
sort_values("Employees",ascending=False)
```

```python
sectors = fortune.groupby("Sector")
```

```python
sectors.size()
```

```python
# gets furst value from all sectors, similarly last()
sectors.first()
```

```python
sectors.groups
```

```python
sectors.get_group('Energy')
```

```python
df = pd.DataFrame()
```

```python
# getting largest Revenue record from each sector
df = df.append([data.nlargest(1,'Revenue') for sector,data in sectors])
```
