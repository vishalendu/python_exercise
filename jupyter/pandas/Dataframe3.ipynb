# Using query 
we have to remove any column names with space

```python
#bond.rename(columns={'Box Office':'BoxOffice','Bond Actor Salary':'BondSalary'},inplace=True)
```

```python
bond = pd.read_csv('data/jamesbond.csv',index_col="Film")
bond['Actor']=bond['Actor'].astype('category')
```

```python
# renaming columns using list comprehension
bond.columns=[column_name.replace(' ','_') for column_name in bond.columns]

# or 
# bond.columns = bond.columns.str.replace(' ','_')
```

```python
bond.head(3)
```

```python
# query is supposed to be faster in searches on large dataframes
bond.query('Actor == "Sean Connery" & Year > 1965')
#bond.query('Actor == "Timothy Dalton" & Budget>60')
```

```python
# Remove nulls from Bond_Actor_Salary
# in python NaN is not equal to NaN
bond.query('Actor == "Sean Connery" & Year > 1965 & Bond_Actor_Salary == Bond_Actor_Salary')
```

```python
bond.query('Actor == "Sean Connery" and Year > 1965 and Bond_Actor_Salary == Bond_Actor_Salary')
```

```python
bond.query('Actor in ["Sean Connery","Roger Moore"] & Year > 1965 & Bond_Actor_Salary == Bond_Actor_Salary')
```

```python
bond['Box_Office'].apply(lambda bo: str(bo) + " Millions!")
```

```python
def critic(row):
    review=''
    actor = row[1]
    budget = row[4]
    if actor in ('Pierce Brosnan','Daniel Craig','Sean Connery'):
        review = review + ' Good'
    elif actor =='Timothy Dalton':
        review = review + ' Bad'
    else:
        review = review + ' Ugly'
    if budget>100:
        review = review + ' ,Costly'
    return review
```

```python
bond['MyReview'] = bond.apply(critic,axis='columns')
```
