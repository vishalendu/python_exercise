## another way to convert dates when loading the csv

```python
emp = pd.read_csv('data/employees.csv',parse_dates=['Start Date','Last Login Time'])
emp.columns=['FirstName','Gender','StartDate','LastLoginTime','Salary','Bonus%',\
              'SeniorMgmt','Team']
emp['SeniorMgmt'] = emp['SeniorMgmt'].astype('bool')
emp['Team'] = emp['Team'].astype('category')
emp['Gender'] = emp['Gender'].astype('category')
```

```python
emp.info()
```

```python
# male management employees sorted by descending salary
emp[(emp['Gender']=='Male') & (emp['SeniorMgmt']==True)].sort_values(by=['Salary'],ascending=False)
```

```python
# check multiple 'OR' combinations using isin method
emp[emp['Team'].isin(['Sales','Distribution','Marketing'])]
# you can even pass another series from a DF into the isin method, to compare between two DFs
```

```python
# getting columns with null values
emp[emp['FirstName'].isnull()]
```

```python
# getting columns with not null values
#emp[emp['FirstName'].isnull()==False]
emp[emp['FirstName'].notnull()]
```

```python
# check series value between any two values (inclusive)
emp[emp['Bonus%'].between(4,8)]
```

```python
emp[emp['StartDate'].between('1991-01-01','1992-01-01')]
```

```python
emp[emp['LastLoginTime'].between('12:00PM','02:30PM')]
```

```python
emp[emp['FirstName'].duplicated()].sort_values(by='FirstName')
# second occurance onwards repeating values are considered as duplicates(keep='First')
# useful if the first entity is the one you want to considered, could be duplicate master data
# emp[emp['FirstName'].duplicated(keep='First')==False].sort_values(by='FirstName')


# last occurance is not considered duplicate (keep='Last') - useful if you have multiple values
# for the same entity and you want to take the last update...
# emp[emp['FirstName'].duplicated(keep='Last')==False].sort_values(by='FirstName')


# Remove all values that have duplicates
# emp[emp['FirstName'].duplicated(keep=False)==False].sort_values(by='FirstName')
```

```python
# remove duplicates from second occurance onwards
emp.drop_duplicates(subset=['FirstName'],keep='first')['FirstName'].count()

# remove duplicates keep the last (latest) one
#emp.drop_duplicates(subset=['FirstName'],keep='last')['FirstName'].count()

# same for keep=False as mentioned above- remove any value that has duplicate

# you can also pass multiple values in subset, then the uniqeness will be 
# computed over multiple columns
emp.drop_duplicates(subset=['FirstName','Gender','Team'],keep='first')
```

Difference between len() and nunique() 
len counts NaN but nunique()  doesnt consider it as valid value, so it doesnt count it

alternatively nunique(dropna=False) <- this will allow counting null as unique value
