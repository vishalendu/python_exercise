```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips_us = tips.groupby(['day','sex']).agg({'size':'sum','total_bill':'mean'}).unstack()
```

```python
tips_us_cp = tips_us.copy()
tips_us_cp.columns = ['__'.join(col).strip() for col in tips_us_cp.columns.values ]
```
