```python
import pandas as pd
```

```python
wk1 = pd.read_csv('data/Restaurant - Week 1 Sales.csv').sort_values('Customer ID')
wk2 = pd.read_csv('data/Restaurant - Week 2 Sales.csv').sort_values('Customer ID')
cust = pd.read_csv('data/Restaurant - Customers.csv')
food = pd.read_csv('data/Restaurant - Foods.csv')
satis = pd.read_csv('data/Restaurant - Week 1 Satisfaction.csv')
```

```python
## Add df one after the other
wk_comb = pd.concat([wk1,wk2],ignore_index=True)
```

```python
wk_comb
```

```python
# inner join by customer id , customers common to both weeks
wk_mrg = wk1.merge(wk2,on='Customer ID',suffixes=['_week1','_week2'],how='inner',indicator=True)
```

```python
wk_mrg
```

```python
wk1_mrg = wk1.merge(wk2,on='Customer ID',suffixes=['_week1','_week2'],how='outer',indicator=True)
```

```python
wk1_mrg['_merge'].value_counts()
```

```python
wk1.merge(food,how='left',on='Food ID').sort_values('Customer ID')
```

```python
wk2.merge(cust,left_on='Customer ID',right_on='ID',how='left').drop('ID',axis=1)
```

```python
wk3 = wk1.copy()
cust1= cust.copy()
cust1 = cust1.set_index('ID')
food1 = food.copy()
food1 = food1.set_index('Food ID')
```

```python
# due to one column and one index, duplicate columns dont come
# result only has one Customer ID column
cust_merg = wk3.merge(cust1,how='left',left_on='Customer ID',right_index=True,sort=True)
```

```python
tot_merg = cust_merg.merge(food1,how = 'left',left_on='Food ID',right_index=True,sort=True)
```
