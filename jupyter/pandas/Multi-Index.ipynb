# Pivot

```python
sale = pd.read_csv('data/salesmen.csv',parse_dates=['Date'])
sale['Salesman'] = sale['Salesman'].astype('category')
```

```python
sale
```

```python
sale.info()
```

```python
sale.pivot(index='Date',columns='Salesman',values='Revenue')
```

# pivot_table()

```python
food = pd.read_csv('data/foods.csv')
for col in ['Gender','City','Frequency','Item']:
    food[col] = food[col].astype('category')
```

```python
food.info()
```

```python
food.pivot_table(values='Spend',index=['Gender','Item'],aggfunc=['count','max','min','mean'])
```

```python
food.pivot_table(values='Spend',index=['Gender','Item'],columns='City',\
                 aggfunc=['count','mean'])
```

```python
import pandas as pd
```

```python
stock = pd.read_csv('http://bit.ly/smallstocks')
```

```python
stock.head()
```

```python
stock.info()
```

```python
# Group by to aggregating 
symb = pd.DataFrame()
symb['Mean'] = stock.groupby('Symbol').Close.mean()
symb['Max'] = stock.groupby('Symbol').Close.max()
symb['Min'] = stock.groupby('Symbol').Close.min()
symb['90p'] = stock.groupby('Symbol').Close.quantile(.90)
symb['25p'] = stock.groupby('Symbol').Close.quantile(.25)
```

```python
symb
```
